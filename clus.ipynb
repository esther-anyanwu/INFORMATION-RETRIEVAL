```python
import pandas as pd
df = pd.read_csv("newstitle (1).csv", encoding = 'utf-8') #Reading the documents to a pandas Dataframe

df.head()
```

```python
# Step 2: Explore the data
df.info()

# Step 3: Data preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

documents = df['overview'].values.astype("U")
```

```python
documents
len(documents)
```

```python
#importing the needed modules

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
#silhouette score is used to get which kvalue is the most appropriate for our cluster.
#the kvalue is the number of cluster that is used in our kmeans

#we are testing a range of values from 2-200, as our k value can't be 1
from sklearn.metrics import silhouette_score

sil = []
kmax = 200
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)


# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(features)
  labels = kmeans.labels_
  sil.append(silhouette_score(features, labels, metric = 'euclidean'))
```

```python
sil
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
k =300
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)

df['cluster'] = model.labels_


```

```python
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=300): {silhouette_score(features, model.labels_, metric="euclidean")}')
```

```python
label=model.predict(features)
label
```

```python
# output the result to a text file.

clusters = df.groupby('cluster')    

for cluster in clusters.groups:
    f = open('cluster'+str(cluster)+ '.csv', 'w', encoding="utf-8") # create csv file
    data = clusters.get_group(cluster)[['title','overview']] # get title and overview columns
    f.write(data.to_csv(index_label='id')) # set index to id
    f.close()

print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :20]: #print out 20 feature terms of each cluster
        print (' %s' % terms[j])
    print('------------')
```
